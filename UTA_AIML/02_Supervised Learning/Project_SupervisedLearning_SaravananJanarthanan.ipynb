**Q 5A. Experiment with various parameters to improve performance of the base model**

**Ans 5A**


```python
# Split the data between train and test in a 75:25 ratio
x_train1, x_test1, y_train1, y_test1 = train_test_split(X_Z_Scaled, Y, test_size=0.25, random_state=4)
KNN_Classifiers1 = KNeighborsClassifier(n_neighbors=7, weights = 'distance')
KNN_Classifiers1.fit( x_train1, y_train1 )

y_predicted_labels1 = KNN_Classifiers1.predict( x_test1 )
KNN_Classifiers1.score( x_test1, y_test1 )
```

```python
# Split the data between train and test in a 85:15 ratio
x_train2, x_test2, y_train2, y_test2 = train_test_split(X, Y, test_size=0.15, random_state=1)
KNN_Classifiers2 = KNeighborsClassifier( n_neighbors=5 )
KNN_Classifiers2.fit( x_train2, y_train2 )

y_predicted_labels2 = KNN_Classifiers2.predict( x_test2 )
KNN_Classifiers2.score( x_test2, y_test2 )
```

```python
# Split the data between train and test in a 70:30 ratio
x_train3, x_test3, y_train3, y_test3 = train_test_split(X_Z_Scaled, Y, test_size=0.3, random_state=3)
KNN_Classifiers3 = KNeighborsClassifier( n_neighbors=7)
# algorithm ‘ball_tree’, ‘kd_tree’, ‘brute’
KNN_Classifiers3.fit( x_train3, y_train3 )

y_predicted_labels3 = KNN_Classifiers3.predict( x_test3 )
KNN_Classifiers3.score( x_test3, y_test3 )
```

```python
# Calculate and plot the confusion matrix for three different data splits
cf_matrix_NS1 = confusion_matrix(y_test1, y_predicted_labels1, labels=["Normal" , "Type_S" ])
cf_matrix_NS2 = confusion_matrix(y_test2, y_predicted_labels2, labels=["Normal" , "Type_S" ])
cf_matrix_NS3 = confusion_matrix(y_test3, y_predicted_labels3, labels=["Normal" , "Type_S" ])

mod_cm_NS1 = pd.DataFrame( cf_matrix_NS1, index = [i for i in ['Normal', 'Type_S' ]],  
                           columns = [ i for i in ['Predict Normal', 'Predict Type_S']])

mod_cm_NS2 = pd.DataFrame( cf_matrix_NS2, index = [i for i in ['Normal', 'Type_S' ]],  
                           columns = [ i for i in ['Predict Normal', 'Predict Type_S']])

mod_cm_NS3 = pd.DataFrame( cf_matrix_NS3, index = [i for i in ['Normal', 'Type_S' ]],  
                           columns = [ i for i in ['Predict Normal', 'Predict Type_S']])

f1,(axn1,axn2,axn3) = plt.subplots(1,3,sharey=False, gridspec_kw={'width_ratios':[0.9,0.9,1]}, figsize=(20,5))

# Configure the three heatmaps for confusion matrix
hm_m1 = sns.heatmap(mod_cm_NS1, annot=True, cmap='Blues', ax=axn1)
hm_m1.set_title('Confusion Matrix (75% training data; neighbour 7)\n Normal and Type_S\n\n');
hm_m1.set_xlabel('\nPredicted Values')
hm_m1.set_ylabel('Actual Values ');

hm_m2 = sns.heatmap(mod_cm_NS2, annot=True, cmap='Blues', ax=axn2)
hm_m2.set_title('Confusion Matrix (85% training data; neighbour 5)\n Normal and Type_S\n\n');
hm_m2.set_xlabel('\nPredicted Values')
hm_m2.set_ylabel('Actual Values ');

hm_m3 = sns.heatmap(mod_cm_NS3, annot=True, cmap='Blues', ax=axn3)
hm_m3.set_title('Confusion Matrix (70% training data; neighbour 7)\n Normal and Type_S\n\n');
hm_m3.set_xlabel('\nPredicted Values')
hm_m3.set_ylabel('Actual Values ');

plt.show()
```

```python
# Calculate the performance scores for neignbours value configured between 1 to 50 for each split data set
scores1 =[]
for k in range(1,50):
    NNH = KNeighborsClassifier(n_neighbors = k, weights = 'distance' )
    NNH.fit(x_train1, y_train1)
    scores1.append(NNH.score(x_test1, y_test1))
    
scores2 =[]
for k in range(1,50):
    NNH = KNeighborsClassifier(n_neighbors = k, weights = 'distance' )
    NNH.fit(x_train2, y_train2)
    scores2.append(NNH.score(x_test2, y_test2))
    
scores3 =[]
for k in range(1,50):
    NNH = KNeighborsClassifier(n_neighbors = k, weights = 'distance' )
    NNH.fit(x_train3, y_train3)
    scores3.append(NNH.score(x_test3, y_test3))    
```

```python
# Plot the performance scores captured in the iterations using neighbour count from 1 to 50
f1, (axp1, axp2, axp3) = plt.subplots(1,3,sharey=False, gridspec_kw={'width_ratios':[0.9,0.9,1]}, figsize=(20,5))

axp1.set(xlabel='\nNeighbours', ylabel='Performance score')
axp2.set(xlabel='\nNeighbours', ylabel='Performance score')
axp3.set(xlabel='\nNeighbours', ylabel='Performance score')

axp1.set_title('Performance score (75% training data; neighbour 1 - 50)\n\n');
axp2.set_title('Performance score (85% training data; neighbour 1 - 50)\n\n');
axp3.set_title('Performance score(70% training data; neighbour 1 - 50)\n\n');

axp1.plot(range(1,50),scores1)
axp2.plot(range(1,50),scores2)
axp3.plot(range(1,50),scores3)

plt.show()
```
