âœ¨ Engineered Features
Amt_to_Balance_Ratio
- Indicates how significant the transaction is relative to the account balance.
- Higher ratios may signal risky behavior or fraud.

Is_Night
- Flags transactions occurring between 10 PM and 6 AM.
- Fraud tends to happen more during odd hours.

High_Value_Transaction
- Binary flag for transactions over $10,000.

Device_Location_Mismatch
- Compares transaction location to the customer's state. A mismatch may indicate suspicious activity.

Customer_Txn_Count
- Approximates customer activity. Newer or less active accounts with large transactions could be red flags.



```python
# Perform Feature Engineering

# Copy original data
df_fe = df_fraud_cleaned.copy()

# Extract hour from 'Transaction_Time'
df_fe['Transaction_Hour'] = pd.to_datetime(df_fe['Transaction_Time'], format='%H:%M:%S').dt.hour


# Create new features

# 1. Transaction amount relative to account balance
df_fe['Amt_to_Balance_Ratio'] = df_fe['Transaction_Amount'] / (df_fe['Account_Balance'] + 1)  # avoid division by zero

# 2. Hour-based features
df_fe['Is_Night'] = df_fe['Transaction_Hour'].apply(lambda x: 1 if x is not None and (x < 6 or x > 22) else 0)

# 3. Transaction Amount category
df_fe['High_Value_Transaction'] = df_fe['Transaction_Amount'].apply(lambda x: 1 if x > 10000 else 0)

# 4. State-device mismatch indicator (assumption-based feature)
df_fe['Device_Location_Mismatch'] = df_fe.apply(
    lambda row: 1 if row['City'] and row['State'] not in row['Transaction_Location'] else 0, axis=1
)

# 5. Transaction Frequency per Customer (proxy using count)
txn_counts = df_fe.groupby('Customer_ID')['Transaction_ID'].count()
df_fe['Customer_Txn_Count'] = df_fe['Customer_ID'].map(txn_counts)

# Preview the engineered features
df_fe_engineered_preview = df_fe[[
    'Transaction_Amount', 'Account_Balance', 'Amt_to_Balance_Ratio',
    'Transaction_Hour', 'Is_Night', 'High_Value_Transaction',
    'State', 'Transaction_Location', 'Device_Location_Mismatch',
    'Customer_ID', 'Customer_Txn_Count', 'Is_Fraud'
]].head(10)

#import ace_tools as tools; tools.display_dataframe_to_user(name="Feature Engineered Sample", dataframe=df_fe_engineered_preview)
```

##### Balance the data through SMOTE #####

```python
# Import required libraries for SMOTE
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split


# Split features and target
X = df_fe_encoded.drop('Is_Fraud', axis=1)
y = df_fe_encoded['Is_Fraud']

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Show the class distribution after SMOTE
resampled_counts = y_resampled.value_counts()
resampled_counts
```

```python
X_resampled.shape, y_resampled.shape
```

##### Train the models and Evaluate them #####

```python
# Import required libraries for model training
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=500, stratify=y_resampled)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


#Train data using Random Forest Model
# Initialize Random Forest Model
rf = RandomForestClassifier(random_state=500)

# Train Random Forest Model
rf.fit(X_train, y_train)

# Predict the trained Random Forest Model
rf_preds = rf.predict(X_test)

# Evaluate the Random Forest Model
rf_report = classification_report(y_test, rf_preds, output_dict=True)

# Load performance summaries
rf_report['model'] = 'Random Forest'
```

```python
#Train data using Logistic Regression Model
# Initialize Logistic Regression Model
lr = LogisticRegression(max_iter=1000, random_state=500)

# Train Logistic Regression Model
lr.fit(X_train_scaled, y_train)

# Predict the trained Logistic Regression Model
lr_preds = lr.predict(X_test_scaled)

# Evaluate the Logistic Regression Model
lr_report = classification_report(y_test, lr_preds, output_dict=True)

# Load performance summaries
lr_report['model'] = 'Logistic Regression'
```

```python
from xgboost import XGBClassifier

# Train XGBoost classifier
# Initialize XGBoost classifier
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=500)

    # Train XGBoost classifier Model
xgb_model.fit(X_train_scaled, y_train)

    # Predict the trained XGBoost classifier Model
y_pred = xgb_model.predict(X_test_scaled)

    # Evaluate the XGBoost classifier Model
xgb_report = classification_report(y_test, y_pred, output_dict=True)

# Load performance summaries
xgb_report['model'] = 'XGBoost classifier'
```

```python
# Train data using Neural Network
    # Initialize or build using NN
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer=Adam(0.001), loss='binary_crossentropy', metrics=['accuracy'])
    # Train Random Forest Model
model.fit(X_train_scaled, y_train, validation_split=0.2, epochs=20, batch_size=32, verbose=1)
    # Predict the trained Random Forest Model
nn_probs = model.predict(X_test_scaled).flatten()
nn_pred = (nn_probs > 0.5).astype(int)
    # Evaluate the Random Forest Model
nn_report = classification_report(y_test, nn_pred, output_dict=True)
nn_report['model'] = 'Neural Network'
```

```python
# Combine reports for comparison
model_reports = pd.DataFrame([{
    'Model': rf_report['model'],
    'Precision (Fraud)': rf_report['1']['precision'],
    'Recall (Fraud)': rf_report['1']['recall'],
    'F1-Score (Fraud)': rf_report['1']['f1-score'],
    'Accuracy': rf_report['accuracy']
}, {
    'Model': lr_report['model'],
    'Precision (Fraud)': lr_report['1']['precision'],
    'Recall (Fraud)': lr_report['1']['recall'],
    'F1-Score (Fraud)': lr_report['1']['f1-score'],
    'Accuracy': lr_report['accuracy']
},  {
    'Model': xgb_report['model'],
    'Precision (Fraud)': xgb_report['1']['precision'],
    'Recall (Fraud)': xgb_report['1']['recall'],
    'F1-Score (Fraud)': xgb_report['1']['f1-score'],
    'Accuracy': xgb_report['accuracy']
},  {
    'Model': nn_report['model'],
    'Precision (Fraud)': nn_report['1']['precision'],
    'Recall (Fraud)': nn_report['1']['recall'],
    'F1-Score (Fraud)': nn_report['1']['f1-score'],
    'Accuracy': nn_report['accuracy']
}
])


model_reports
```
